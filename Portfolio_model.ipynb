```python
stock_symbol = []
for s in slist:
  search_results = investpy.search_quotes(text=s, 
                                        countries=['india'],
                                        n_results=5)
  print(search_results[0].symbol)
  stock_symbol.append(search_results[0].symbol)
print(len(stock_symbol))
```

```python
# stock_symbo
def get_hist_quotes(sname):
  df = investpy.get_stock_historical_data(stock=sname, country='india', 
                                        from_date=from_date, to_date=to_date, )['Close']
  return df

```

```python
df = pd.DataFrame()
df[stock_symbol[0]] = get_hist_quotes(stock_symbol[0])
for s in stock_symbol[1:]:
  temp = pd.DataFrame()
  temp[s] = get_hist_quotes(s)
  df = pd.merge(left=df, right=temp, how='left', left_index=True, right_index=True, )
```

```python
pd.set_option('display.max_columns', None)
print(df.shape)
df.tail(10)
```

```python
### removing HDFL
# hdfl = df.HDFL
sbil = df.SBIL
df = df.drop(['HDFL', 'SBIL'], axis=1)
```

```python
# df.to_csv('/content/drive/MyDrive/Colab Notebooks/Portfolio/nifty_stocks.csv')
```

```python
daily_return = df.pct_change(1).dropna()
```

```python
daily_return
```

```python
daily_return.describe().transpose()
```

```python
plt.figure(figsize=(30, 8))
sns.heatmap(daily_return.corr(), annot=True, )
```

```python
cov_matrix_annual = daily_return.cov() * 252
print(cov_matrix_annual.shape)
```

```python
###Expected portfolio variance= WT * (Covariance Matrix) * W
import numpy as np
weights = np.array([.02]*48)
print(weights)
port_variance = np.dot(weights.T, np.dot(cov_matrix_annual, weights))
port_variance
```

```python
port_volatility = np.sqrt(port_variance)
port_volatility
```

```python
portfolioSimpleAnnualReturn = np.sum(daily_return.mean()*weights) * 252
portfolioSimpleAnnualReturn
```

```python
percent_var = str(round(port_variance, 2) * 100) + '%'
percent_vols = str(round(port_volatility, 2) * 100) + '%'
percent_ret = str(round(portfolioSimpleAnnualReturn, 2)*100)+'%'

print("Expected annual return : "+ percent_ret)
print('Annual volatility/standard deviation/risk : '+percent_vols)
print('Annual variance : '+percent_var)
```

# Portfolio opt

```python
!pip install -q PyPortfolioOpt
```

```python
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models
from pypfopt import expected_returns
```

# New Section

```python
mu = expected_returns.mean_historical_return(df)#returns.mean() * 252
S = risk_models.sample_cov(df) #Get the sample covariance matrix
```

```python
ef = EfficientFrontier(mu, S)
weights = ef.max_sharpe() 
#Maximize the Sharpe ratio, and get the raw weights
cleaned_weights = ef.clean_weights() 
print(cleaned_weights) 
#Note the weights may have some rounding error, meaning they may not add up exactly to 1 but should be close
ef.portfolio_performance(verbose=True)
```

```python
!pip install -q pulp
```

```python
from pypfopt.discrete_allocation import DiscreteAllocation, get_latest_prices
latest_prices = get_latest_prices(df)
weights = cleaned_weights 
da = DiscreteAllocation(weights, latest_prices, total_portfolio_value=50000)
allocation, leftover = da.lp_portfolio()
print("Discrete allocation:", allocation)
print("Funds remaining: ${:.2f}".format(leftover))
```
